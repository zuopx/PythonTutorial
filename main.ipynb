```python
import matplotlib.pyplot as plt
```

```python
plt.rcParams['font.sans-serif']=['WenQuanYi Micro Hei']
plt.rcParams['axes.unicode_minus'] = False
```

```python
ind = (0, 1)
data = (170, 96)
width = 0.5
myxlim = (-0.5, 1.5)
myylim = (0, 200)
xticklabels = ('本发明的模型', '现有模型')
ylabel = '信息传播范围的大小'
```

```python
fig, ax = plt.subplots(dpi=100, figsize=(8, 6))
ax.bar(ind, data, width, color='w', edgecolor='k', hatch='/')
ax.set_xlim(*myxlim)
ax.set_ylim(*myylim)
ax.set_xticks(ind)
ax.set_xticklabels(xticklabels, fontsize=20)
ax.set_ylabel(ylabel, fontsize=20)
```

```python
fig.tight_layout()
```
